```python
usuarios = []

while True:
    nome = input("Digite seu nome: ")
    email = input("Digite seu E-mail: ")
    senha = input("Digite sua senha: ")
    usuario = {"nome": nome, "email": email, "senha": senha}
    usuarios.append(usuario)

    pergunta = int(input("Deseja continuar? 1-Sim 2-Não "))
    if pergunta < 1 or pergunta > 2 :
        print("Erro. Digite 1 para Sim e 2 para Não")
        pergunta = int(input("Deseja continuar? 1-Sim 2-Não "))
    if pergunta == 1 :
        continue
    if pergunta == 2 :
        print("Programa encerrado!")
        break

print("\nUsuários cadastrados:")
for usuario in usuarios:
    print(f"Nome: {usuario['nome']} | Email: {usuario['email']}")
```
